```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import glob
```

```python
data = open("testcreation/tests/lab13.out").read().split(' ')[:-1]
plt.plot([float(i) for i in data])
plt.show()
```

```python
to_test = glob.glob("./testcreation/tests/lab*.out")
to_test += glob.glob("./testcreation/tests/ran*.out")
sums = []
names = []
for i in to_test:
    data = open(i).read().split(' ')[:-1]
    names.append(i[len('./testcreation/tests/'):])
    sums.append(sum([float(i) for i in data]))
d = np.array([names, sums]);
d = np.transpose(d);
df = pd.DataFrame(d, columns =['names', 'time'])
```

```python
df = df.sort_values(by=['time'])
```
